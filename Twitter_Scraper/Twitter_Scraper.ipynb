```python
import datetime
import tweepy
import pandas as pd
import re
import json
import time
```

# Load Query CSV
- Jacob: jacob.csv
- Winston: winston.csv
- Niyi: niyi.csv

```python
# Replace your name here in lowercase with the one that matches your csv
name = ''
```

```python
csv = 'Queries/'+ name + '_queries.csv'
```

```python
query_df = pd.read_csv(csv)
query_df['Query'] = query_df['Query'].apply(lambda x: x.strip())
query_df['Date'] = query_df['Date'].apply(lambda x: x.strip())
query_df
```

```python
format_str = '%m/%d/%Y'
```

```python
query_df['Query'] = query_df['Query'].apply(lambda x: re.sub("[\(\[].*?[\)\]]", "", x)) 
query_df['From_Date'] = query_df['Date'].apply(lambda date_str: datetime.datetime.strptime(date_str, format_str))
query_df['To_Date'] = query_df['From_Date'].apply(lambda date_obj: date_obj + pd.Timedelta(1, 'days'))
```

```python
query_df['From_Date'] = query_df['From_Date'].apply(lambda x: x.strftime("%Y%m%d%H%M"))
query_df['To_Date'] = query_df['To_Date'].apply(lambda x: x.strftime("%Y%m%d%H%M")) 
```

# Search Tweets
Saved Documentation:
- https://developer.twitter.com/en/docs/twitter-api/enterprise/rules-and-filtering/operators-by-product
- twitter.com/anyuser/status/tweet_id allows searching for tweets by tweet id

```python
# Construct the API instance
# Replace these with your API Key and API Secret Key
api_key = 'REPLACE'
api_secret_key = 'REPLACE'
auth = tweepy.AppAuthHandler(api_key, api_secret_key)
tweepy_api = tweepy.API(auth)
```

```python
# Replace this label with the dev envrioment label you created for the full archive search
dev_environment_label = 'REPLACE'
query_operators = 'lang:en'
```

```python
# This list holds the tweets search by tweepy
tweet_json_list = []
```

```python
# Loop over each query and append the output to the list
for i in range(40, len(query_df)):
    # Wait for rate limit to refresh half way through searching queries
    if i == 25:
        time.sleep(120)
    returned_tweets = tweepy_api.search_full_archive(dev_environment_label, query_df['Query'][i] + ' ' + query_operators, fromDate=query_df['From_Date'][i], toDate=query_df['To_Date'][i], maxResults=100)
    for tweet in returned_tweets:
        tweet_json_list.append(tweet._json)
    

    
```
